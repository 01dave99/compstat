```python
import numpy as np
import scipy.stats as sc
import matplotlib.pyplot as plt
```

```python

Ax = np.array([0, 26981, 39534, 74242, 68635, 117449])
Ais = np.array([34583, 44958, 30087, 39770, 26629, 38733])
c = np.array([0, 1, 2, 3, 4, 5])

res = sc.linregress(c, Ax/Ais)
line = np.array([res.slope * i + res.intercept for i in c])


plt.scatter(c, Ax/Ais, c="red")
plt.plot(c, line, label=f"y = {res.slope}x +  {res.intercept}")
plt.legend()
plt.show()
```
